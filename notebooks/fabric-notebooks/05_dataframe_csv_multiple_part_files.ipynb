```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DatapurProgram").getOrCreate()
```

```python
# Unity Catalog volume path
rootPath = "abfss://shamas_ws@onelake.dfs.fabric.microsoft.com/test_Lakehouse.Lakehouse/Files/client_output_data/csv/"
folderPath = rootPath + "students_csv"  # folder for CSV dataset

data = [
    (1, "Alice", 21),
    (2, "Bob", 22),
    (3, "Charlie", 23),
    (4, "David", 24)
]

columns = ["id", "name", "age"]

df_student = spark.createDataFrame(data, columns)
#display(df_student)

df_student.write \
    .mode("overwrite") \
    .option("header", "true") \
    .csv(folderPath)
print("CSV file written to " + folderPath)
```

```python
df_csv = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(folderPath)

display(df_csv)
```

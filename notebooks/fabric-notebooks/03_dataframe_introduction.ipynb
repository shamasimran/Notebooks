```python
from pyspark.sql import SparkSession
# Create a Spark session
spark = SparkSession.builder.appName("BigEntities").getOrCreate()
```

```python
# Example array (list of tuples)
data = [
    (1, "Alice", 34),
    (2, "Bob", 45),
    (3, "Charlie", 29)
]

# Column names
columns = ["ID", "Name", "Age"]

# Create DataFrame from array
df = spark.createDataFrame(data, columns)

# Show DataFrame
display(df)
```

```python
data = [("Alice", 34), ("Bob", 45), ("Charlie", 29)]
df = spark.createDataFrame(data)
#df.show()

display(df)
```

```python
#df.printSchema()
schema = df.schema
display(schema)
```

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType

student_schema = StructType([StructField('ID', LongType(), False), StructField('Name', StringType(), True), StructField('Age', LongType(), True)])

student_data = [
        (1, "Alice", 34), 
        (2, "Bob", 45), 
        (3, "Charlie", 29),
        (4, "Shamas", 40),
        (5, "Nuzha", None)
        ]

df_student = spark.createDataFrame(student_data, student_schema)
df_student.show()
```

```python
from pyspark.sql.types import StructType, StructField, LongType, StringType

student_schema = StructType([StructField('ID', LongType(), False), StructField('Name', StringType(), True), StructField('Age', LongType(), True)])

student_data = [
        (1, "Alice", 34), 
        (2, "Bob", 45), 
        (3, "Charlie", 29),
        (4, "Shamas", 40),
        (5, "Nuzha", None),
        (None, "Nuzha", None),
        ]

df_student = spark.createDataFrame(student_data, student_schema,None, False)
df_student.show()
```
